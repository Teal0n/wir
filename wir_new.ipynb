## Μέθοδοι
* get_image_files: Αυτή η συνάρτηση χρησιμοποιείται για να λάβει τα αρχεία
εικόνων από τον καθορισμένο φάκελο. Χρησιμοποιείται για την ανάκτηση των εικόνων που θα χρησιμοποιηθούν στην εκπαίδευση και τη δοκιμή του μοντέλου.

* parent_label: Αυτή η συνάρτηση χρησιμοποιείται για να αντιστοιχίσει τις ετικέτες κατηγορίας των εικόνων με βάση τον γονικό φάκελο της κάθε εικόνας.

* DataBlock: Αυτή η κλάση χρησιμοποιείται για τον ορισμό του data block, το οποίο καθορίζει τη δομή και τις διαδικασίες που απαιτούνται για τη δημιουργία του dataset. Συγκεκριμένα, καθορίζει τα blocks των εικόνων και των ετικετών, τον τρόπο ανάκτησης των αρχείων εικόνων και την αντιστοίχιση των ετικετών, τον τρόπο διαίρεσης του dataset σε train και validation sets, τους μετασχηματισμούς που πρέπει να εφαρμοστούν στις εικόνες, όπως η αλλαγή μεγέθους των εικονών, και την επαύξηση δεδομένων.

* dataloaders: Αυτή η συνάρτηση δημιουργεί τους dataloaders, δηλαδή τα αντικείμενα που φορτώνουν τα δεδομένα (εικόνες και ετικέτες) και τα επεξεργάζονται για την εκπαίδευση και τη δοκιμή του μοντέλου.

* vision_learner: Αυτή η κλάση δημιουργεί έναν learner που είναι υπεύθυνος για την εκπαίδευση του μοντέλου. Ορίζει τον τύπο του μοντέλου, τις μετρικές που θα χρησιμοποιηθούν για την αξιολόγηση του μοντέλου, και τα dataloaders που πρέπει να χρησιμοποιηθούν.

* fine_tune: Αυτή η συνάρτηση εκπαιδεύει το μοντέλο με την τεχνική fine-tuning. Πραγματοποιεί μια αρχική εκπαίδευση του μοντέλου στο σύνολο δεδομένων και στη συνέχεια εφαρμόζει τη μέθοδο early stopping για να σταματήσει την εκπαίδευση αν δεν παρουσιαστεί βελτίωση στην απόδοση.

* show_results: Αυτή η μέθοδος εμφανίζει τα αποτελέσματα της ταξινόμησης για ορισμένα παραδείγματα εικόνων.

* ClassificationInterpretation: Αυτή η κλάση παρέχει λειτουργίες για τον ερμηνευτή των αποτελεσμάτων της ταξινόμησης. Επιτρέπει την προβολή των κορυφαίων απωλειών (top losses) και την απεικόνιση του πίνακα σύγχυσης (confusion matrix).

* export: Αυτή η μέθοδος χρησιμοποιείται για την εξαγωγή του μοντέλου σε ένα αρχείο.

## Import libraries

```python
from fastai.vision.all import *
import warnings

# Ignore annoying warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

## Preprocessing

### Load

```python
# Path to the main folder containing subfolders of weather images
path = Path('/content/drive/MyDrive/Colab_Dataset/dataset')
```

### Datablock & Dataloaders

```python
# Create the data block
dblock = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    get_y=parent_label,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    item_tfms=Resize(224),
    batch_tfms=aug_transforms()
)

# Create the dataloaders
dls = dblock.dataloaders(path)
```

### Batch

```python
# Show a batch of data
dls.show_batch()
```

## Classification Task

```python
# Create the model
learn = vision_learner(dls, resnet18, metrics=accuracy)
```

### Fine-Tuning

```python
# Model training
learn.fine_tune(10, cbs=EarlyStoppingCallback(patience=3))
```

### Predictions

```python
# Show results
learn.show_results()
```

### Metrics

```python
# Model analysis
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_top_losses(9)
```

```python
interp.plot_confusion_matrix()
```

### Save

```python
# Save the model
learn.export('/content/drive/MyDrive/Colab_Models/model1810.pkl')
```
